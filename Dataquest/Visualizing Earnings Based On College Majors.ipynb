```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full_time')
```

There is no relation between the full time jobs and the median salary.

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. ShareWomen')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs. Men')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs. Women')
```

Men have a bigger salary than women, but the chart can be misleading. It would be worth to take a look closer.

```python
ax = recent_grads.plot(x='Total', y='Median', kind='scatter', c='ShareWomen',
      s= 150, title='Median vs Total', colormap='bwr', figsize=(10,5))
ax.set_xlabel('Total')
```

The median salary is higher in majors where the majority of students are men (shown by the blue dots)

```python
recent_grads.loc[recent_grads['Total'] > recent_grads['Total'].mean(),
                 'Popular'] = 1
recent_grads.loc[recent_grads['Total'] <= recent_grads['Total'].mean(),
                 'Popular'] = 0

ax = recent_grads.plot(x='Major_code', y = 'Median', kind = 'scatter',
                       c='Popular',  figsize=(12,8), colormap='bwr',
                       s=180)
ax.set_xlabel('Major code')
ax.set_title('Median vs Major code')
```

There seems to be no relation between popular majors and the median income
