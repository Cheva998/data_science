```python
autos['odometer_km'] = autos.odometer.str.replace('km','')
autos['odometer_km'] = autos.odometer_km.str.replace(',','')
autos['odometer_km'] = autos['odometer_km'].astype(int)
```

```python
autos['price_usd'] = autos.price.str.replace('$','')
autos['price_usd'] = autos.price_usd.str.replace(',','')
autos['price_usd'] = autos['price_usd'].astype(int)
```

### Remove the values before 1960 because older vehicles may be a misstype. The right thing to do was to review the name of the vehicles and check if that model was available at that year. But since I didn't have the time, I decided to make a Salomonic decision 
### Remove the values after 2016 because the ad was created in 2016, there can't be vehicles from the future. The right thing to do was to check if the registration year matches the model and the ad creation date.

```python
autos.groupby(by='brand')[['odometer_km', 'price_usd']].mean()
```

```python
brands_counts = autos.brand.value_counts()
```

### I choose the top 20 brands to make the analysis

```python
brands_price = {}
brands_mileage = {}
for i in brands_counts.index[:20]:
    brands_mileage[i] = brands_mileage.get(i, \
       autos[autos['brand'] == i]['odometer_km'].mean())
    brands_price[i] = brands_price.get(i, \
      autos[autos['brand'] == i]['price_usd'].mean())
# brands_counts.index[:20]
```

```python
brands_price
```

### We see that the skoda is the most expensive ... on average

```python
price_serie = pd.Series(brands_price)
mileage_serie = pd.Series(brands_mileage)
```
